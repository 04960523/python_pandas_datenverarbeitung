```
# string manipulation function that generates email addresses on demand as alternative of CA fast data masker
import pandas as pd
import sys
def cutstring(oldstr):
    newstr = oldstr[:4] +oldstr[5:]
    return newstr
def front_back(string):
    if len(string) == 1:
        return string
    return string[-1] + string[1:-1] + string[0]  # swap the first and last character in a given string  
print(sys.version)
df = pd.read_csv('test4.csv',encoding='windows-1252',delimiter=';')
df1= pd.DataFrame(df, columns=['ID', 'Geburtstag', 'Email_Addresse', 'Ansprechpartner (Vorname)','Insurance','Letzter Rechnungsbetrag (€)','Ansprechpartner (Nachname)','Ansprechpartner','Ansprechpartner (Telefon)','private_insurance'])

i=0
for x in df1['Email_Addresse']:
    z=cutstring(x)
    y=front_back(z)
   # print (y)
    df1.at[i,'Email_Addresse']=y
    i=i+1
 

df1.to_csv('test5.csv',sep=';',index=0)   
    
```

```
from numpy import *
import math
import matplotlib.pyplot as plt
import matplotlib


x1 = linspace(0, 9, 9)
a = 1-exp(-x1*x1/2)
b = 1-exp(-x1*x1/8)
c = 1-exp(-x1*x1/18)
d = 1-exp(-x1*x1/12.5)

fig = plt.figure()


fig.subplots_adjust(top=0.85)



fig2 = plt.figure()
bx = fig2.add_subplot(111)
bx.set_title('wie verändert sich Anonymiesierungsgrad θ mit σ = 1, 2, 3')

bx.set_xlabel('Δx')
bx.set_ylabel('θ')
bx.plot(x1, a, 'r') # sigma=1
bx.plot(x1, b, 'b') # 2
bx.plot(x1, c, 'g') # 3
bx.plot(x1, d, 'c') # 2.5

bx.text(2, 1, ' σ=1')
bx.text(4, 1, ' σ=2')
bx.text(8, 0.9, ' σ=3')
bx.set_xlim([0,10])


plt.show()
```

```
from numpy import *
import math
import matplotlib.pyplot as plt
import matplotlib

x = linspace(0, 300, 400)
a = 1-exp(-0.5/x/x)
b = 1-exp(-50/x/x)
c = 1-exp(-5000/x/x)
d = 1-exp(-125000/x/x)


fig = plt.figure()
ax = fig.add_subplot(111)

fig.subplots_adjust(top=0.85)
ax.set_title('wie verändert sich Anonymiesierungsgrad θ mit vorgegebene Δx=1, 10, 100, 500')

ax.set_xlabel('σ')
ax.set_ylabel('θ')
ax.plot(x, a, 'r') # delta x = 1 
ax.plot(x, b, 'b') # delta x = 10 
ax.plot(x, c, 'g') # delta x = 100
ax.plot(x, d, 'c') # delta x = 100
ax.text(-10, 0.2, 'Δx=1')
ax.text(25, 0.1, 'Δx=10')
ax.text(200, 0.2, 'Δx=100')
ax.text(250, 0.9, 'Δx=500')

plt.show()
```
